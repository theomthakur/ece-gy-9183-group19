```python
# Data loading
def load_test_dataset(dataset_dir):
    images = []
    annotations = []
    for img_path in Path(dataset_dir).glob("*.jpg"):
        ann_path = img_path.with_suffix(".txt")
        if ann_path.exists():
            images.append(str(img_path))
            with open(ann_path, "r") as f:
                anns = [line.strip().split() for line in f if line.strip()]
                anns = [[int(float(a[0])), float(a[1]), float(a[2]), float(a[3]), float(a[4])] for a in anns]
                annotations.append(anns)
    return images, annotations

test_images, test_annotations = load_test_dataset(TEST_DATASET_DIR)

# Evaluate general metrics
def compute_map(predictions, annotations, iou_threshold=0.5):
    aps = []
    for cls in range(len(CLASSES)):
        true_positives, false_positives, false_negatives = 0, 0, 0
        scores = []
        for pred, ann in zip(predictions, annotations):
            pred_boxes = [p for p in pred.boxes.data.cpu().numpy() if int(p[5]) == cls]
            true_boxes = [a for a in ann if a[0] == cls]
            matched = []
            for pb in pred_boxes:
                max_iou, best_idx = 0, -1
                for i, tb in enumerate(true_boxes):
                    if i in matched:
                        continue
                    iou = calculate_iou(pb[:4], tb[1:])
                    if iou > max_iou:
                        max_iou, best_idx = iou, i
                if max_iou >= iou_threshold:
                    true_positives += 1
                    matched.append(best_idx)
                else:
                    false_positives += 1
                scores.append(pb[4])
            false_negatives += len(true_boxes) - len(matched)
        if true_positives + false_positives == 0 or true_positives + false_negatives == 0:
            aps.append(0)
            continue
        precision = true_positives / (true_positives + false_positives)
        recall = true_positives / (true_positives + false_negatives)
        aps.append(precision * recall)
    return np.mean(aps)

def calculate_iou(box1, box2):
    x1, y1, w1, h1 = box1
    x2, y2, w2, h2 = box2
    x1_min, x1_max = x1 - w1/2, x1 + w1/2
    y1_min, y1_max = y1 - h1/2, y1 + h1/2
    x2_min, x2_max = x2 - w2/2, x2 + w2/2
    y2_min, y2_max = y2 - h2/2, y2 + h2/2
    inter_xmin = max(x1_min, x2_min)
    inter_ymin = max(y1_min, y2_min)
    inter_xmax = min(x1_max, x2_max)
    inter_ymax = min(y1_max, y2_max)
    inter_area = max(0, inter_xmax - inter_xmin) * max(0, inter_ymax - inter_ymin)
    union_area = w1 * h1 + w2 * h2 - inter_area
    return inter_area / union_area if union_area > 0 else 0
```

```python
# Template-based tests
def compose_image(food_path, bg_path=None, extra_path=None):
    food = Image.open(food_path).convert("RGBA")
    if bg_path:
        bg = Image.open(bg_path).convert("RGBA")
    else:
        bg = Image.new("RGBA", food.size, (255, 255, 255, 255))
    bg_w, bg_h = bg.size
    y_offset = int(bg_h * 0.05)
    food_scale = 0.5
    food = food.resize((int(bg_w * food_scale), int(bg_h * food_scale)))
    fd_w, fd_h = food.size
    if extra_path:
        extra_scale = 0.35
        extra = Image.open(extra_path).convert("RGBA")
        extra = extra.resize((int(bg_w * extra_scale), int(bg_h * extra_scale)))
        ex_w, ex_h = extra.size
        bg.paste(extra, (bg_w - ex_w, bg_h - ex_h - y_offset), extra)
    bg.paste(food, ((bg_w - fd_w) // 2, bg_h - fd_h - y_offset), food)
    return bg.convert("RGB")

# Test robustness to background and extra items
template_tests = {
    'original': compose_image(f"{TEMPLATE_DIR}/objects/class1/001.png"),
    'bg1_extra1': compose_image(f"{TEMPLATE_DIR}/objects/class1/001.png", f"{TEMPLATE_DIR}/background/001.jpg", f"{TEMPLATE_DIR}/extras/spoon.png"),
    'bg2_extra2': compose_image(f"{TEMPLATE_DIR}/objects/class1/001.png", f"{TEMPLATE_DIR}/background/002.jpg", f"{TEMPLATE_DIR}/extras/fork.png"),
    'same_class': compose_image(f"{TEMPLATE_DIR}/objects/class1/002.png", f"{TEMPLATE_DIR}/background/001.jpg"),
    'diff_class': compose_image(f"{TEMPLATE_DIR}/objects/class2/001.png", f"{TEMPLATE_DIR}/background/001.jpg")
}

fig, axes = plt.subplots(2, 5, figsize=(14, 6))
for i, (key, img) in enumerate(template_tests.items()):
    results = model(img)
    pred_classes = [int(p[5]) for p in results[0].boxes.data.cpu().numpy()]
    pred_label = CLASSES[pred_classes[0]] if pred_classes else "None"
    image_np = np.array(img.resize((224, 224))).astype(dtype=np.float32) / 255.0
    cam = EigenCAM(model, target_layers=[model.model.model[-1]], task='od')
    grayscale_cam = cam(image_np)[0, :, :]
    vis = show_cam_on_image(image_np, grayscale_cam, use_rgb=True)
    axes[0, i].imshow(img.resize((224, 224)))
    axes[0, i].set_title(f"{key}\nPredicted: {pred_label}")
    axes[0, i].axis("off")
    axes[1, i].imshow(vis)
    axes[1, i].axis("off")
plt.tight_layout()
plt.show()
```

```python
# Evaluate on slices of interest
slice_images = [str(p) for p in Path(SLICE_DIR).glob("*.jpg")]
slice_results = [model(img) for img in slice_images]
slice_correct = 0
for res, img_path in zip(slice_results, slice_images):
    ann_path = Path(img_path).with_suffix(".txt")
    if ann_path.exists():
        with open(ann_path, "r") as f:
            true_classes = [int(float(line.split()[0])) for line in f if line.strip()]
        pred_classes = [int(p[5]) for p in res[0].boxes.data.cpu().numpy()]
        if set(pred_classes).issuperset(true_classes):
            slice_correct += 1
print(f"Accuracy on slice of interest: {slice_correct / len(slice_images):.2f}")

# Evaluate on known failure modes
failure_images = [str(p) for p in Path(FAILURE_DIR).glob("*.jpg")]
failure_results = [model(img) for img in failure_images]
failure_correct = 0
for res, img_path in zip(failure_results, failure_images):
    ann_path = Path(img_path).with_suffix(".txt")
    if ann_path.exists():
        with open(ann_path, "r") as f:
            true_classes = [int(float(line.split()[0])) for line in f if line.strip()]
        pred_classes = [int(p[5]) for p in res[0].boxes.data.cpu().numpy()]
        if set(pred_classes).issuperset(true_classes):
            failure_correct += 1
print(f"Accuracy on failure modes: {failure_correct / len(failure_images):.2f}")
```
